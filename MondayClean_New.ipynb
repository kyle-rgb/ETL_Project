```python
season_df = pd.read_csv("2017-18/17-18-cover.csv")
season_df.loc[13, "Team"] = "Oklahoma City"
season_df.loc[3, "Team"] = "L.A. Clippers"
season_df.loc[27, "Team"] = "L.A. Lakers"
```

```python
season_df
```

```python
new_df = merged_df.merge(season_df, on='Team')
```

```python
total_df = new_df.loc[:, ["Team", "Location", "Win", "Loss", "Tie","Cover %", "PercentBet_Spread", "Average_Line_Spread",
               "Average_Odds_Spread", "PTS", "FG_PCT", "FT_PCT", "FG3_PCT"]]
```

```python
total_df.groupby('Team').mean().sort_values('PercentBet_Spread', ascending=False)
```

```python
series_diff = (total_df.groupby('Team').mean()["Cover %"] - total_df.groupby('Team').mean()["PercentBet_Spread"])
```

```python
grouped_df = total_df.groupby('Team').mean().sort_values('Cover %', ascending=False)
```

```python
grouped_df=grouped_df.reset_index()
```

```python
new_grouped_df = grouped_df.merge((series_diff.reset_index()), on="Team")
```

```python
new_grouped_df.rename(columns={0: "Cover% - PercentBetSpread"}, inplace=True)
```
