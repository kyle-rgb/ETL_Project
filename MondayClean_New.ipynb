# Further Analysis

```python
mg_df= vegas_df.merge(df, left_on=["GameId", "TeamId"], right_on=["GAME_ID", "TEAM_ID"])
```

```python
mg_df2 = vegas_df2.merge(df2, left_on=["GameId", "TeamId"], right_on=["GAME_ID", "TEAM_ID"])
```

```python
mg_df.set_index("Date", inplace=True)
```

```python
mg_df2.set_index("Date", inplace=True)
```

```python
mg_dfa = mg_df.copy()
```

```python
mg_dfa = mg_dfa.reset_index()
```

```python
mg_df2 = mg_df2.reset_index()
```

```python
for i, tup in mg_dfa.iterrows():
    y = tup["Spread"]
    mg_dfa.loc[i, "Spread_Result"] = -y
```

```python
for i, tup in mg_df2.iterrows():
    y = tup["Spread"]
    mg_df2.loc[i, "Spread_Result"] = -y
```

```python
for i, row in mg_dfa.iterrows():
    if row["Spread_Result"] > row["Average_Line_Spread"]:
        mg_dfa.loc[i, "Spread_Wins"] = -1
    elif row["Spread_Result"] == row["Average_Line_Spread"]:
        mg_dfa.loc[i, "Spread_Wins"] = 0
    else:
        mg_dfa.loc[i, "Spread_Wins"] = 1   
    
```

```python
for i, row in mg_df2.iterrows():
    if row["Spread_Result"] > row["Average_Line_Spread"]:
        mg_df2.loc[i, "Spread_Wins"] = -1
    elif row["Spread_Result"] == row["Average_Line_Spread"]:
        mg_df2.loc[i, "Spread_Wins"] = 0
    else:
        mg_df2.loc[i, "Spread_Wins"] = 1   
```

```python
mg_dfa.drop("Spread", axis=1, inplace=True)
```

```python
mg_df2.drop("Spread", axis=1, inplace=True)
```

```python
# HOME / AWAY Splits
main_agg = "mean"

full_df.groupby(["Team", "Season", "Location"]).agg({"PercentBet_Spread": main_agg, "Average_Line_Spread": main_agg,
                                                             "PTS": main_agg, "PTS_QTR1": main_agg, "PTS_QTR2": main_agg, "PTS_QTR3": main_agg, "PTS_QTR4": main_agg,
                                                             "FG_PCT": main_agg, "FT_PCT": main_agg, "FG3_PCT": main_agg, "AST": main_agg, "REB": main_agg, "TOV": main_agg,
                                                             "Spread_Result": main_agg, "Spread_Wins": 'sum', "Season": main_agg})
```

```python
# Team by Team
main_agg = "mean"

full_df.groupby(["Team", "Season"]).agg({"PercentBet_Spread": main_agg, "Average_Line_Spread": main_agg,
                                                             "PTS": main_agg, "PTS_QTR1": main_agg, "PTS_QTR2": main_agg, "PTS_QTR3": main_agg, "PTS_QTR4": main_agg,
                                                             "FG_PCT": main_agg, "FT_PCT": main_agg, "FG3_PCT": main_agg, "AST": main_agg, "REB": main_agg, "TOV": main_agg,
                                                             "Spread_Result": main_agg, "Spread_Wins": 'sum', "Season": main_agg})
```

```python
# Spread Type Splits
main_agg = "mean"

full_df.groupby(["spread_status", "Season"]).agg({"PercentBet_Spread": main_agg, "Average_Line_Spread": main_agg,
                                                             "PTS": main_agg, "PTS_QTR1": main_agg, "PTS_QTR2": main_agg, "PTS_QTR3": main_agg, "PTS_QTR4": main_agg,
                                                             "FG_PCT": main_agg, "FT_PCT": main_agg, "FG3_PCT": main_agg, "AST": main_agg, "REB": main_agg, "TOV": main_agg,
                                                             "Spread_Result": main_agg, "Spread_Wins": 'sum', "Season": main_agg})
```

```python
# Spread_status per month
full_df.set_index("Date").groupby(["spread_status", pd.Grouper(freq="1M")]).agg({"PercentBet_Spread": main_agg, "Average_Line_Spread": main_agg,
                                                             "PTS": main_agg, "PTS_QTR1": main_agg, "PTS_QTR2": main_agg, "PTS_QTR3": main_agg, "PTS_QTR4": main_agg,
                                                             "FG_PCT": main_agg, "FT_PCT": main_agg, "FG3_PCT": main_agg, "AST": main_agg, "REB": main_agg, "TOV": main_agg,
                                                             "Spread_Result": main_agg, "Spread_Wins": 'sum', "Season": main_agg})
```
