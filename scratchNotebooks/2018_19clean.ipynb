```python
import pandas as pd
import numpy as np
import os 
```

```python
path18 = "2018-19/vegas.csv"
path2 = "2018-19/raw_scores.csv"
vegas19_df = pd.read_csv(path18, encoding="utf-8")
scores19_df = pd.read_csv(path2, encoding="utf-8")
vegas19_df.head()
```

```python
vegasred19_df = vegas19_df[["Team","Location", "TeamId", "GameId",
                        "PercentBet_Spread", "Average_Line_Spread", "Spread", "Result"]]
vegasred19_df
```

```python
scores19_df.head()
```

```python
redscores19_df = scores19_df[["TEAM_CITY_NAME", "TEAM_ID", "GAME_ID", "PTS",]]
redscores19_df
```

```python
scoresred19_df = redscores19_df.rename(columns={"TEAM_CITY_NAME":"Team", "TEAM_ID":"TeamId",
                                               "GAME_ID":"GameId", "PTS":"Pts"})
scoresred19_df
```

```python
merge_df = pd.merge(vegasred19_df, scoresred19_df, on=["TeamId", "GameId"])
merge_df
```

```python
grouped_df = vegasred19_df.groupby(['GameId'])
print(grouped_df)
```

```python
grouped_df.count().head()
```

```python
hawks_df = vegasred19_df.loc[vegasred19_df["Team"] == "Atlanta"]
hawks_df
```

```python
comp_df = merge_df[["Team_x", "Location", "TeamId", "GameId", "PercentBet_Spread", "Average_Line_Spread", "Spread", "Result", "Pts"]]
comp_df
```

```python
final_df = comp_df.rename(columns= {"Team_x":"Team",
                                    "Location": "Location",
                                   "TeamId":"Team Id",
                                   "GameId":"Game Id",
                                   "PercentBet_Spread":"% Bet",
                                   "Average_Line_Spread": "Avg Spread",
                                   "Spread": "Actual Spread",
                                   "Result":"Result",
                                   "Pts": "Pts"})
final_df
```

```python
Utah_df = final_df.loc[final_df["Team"] == "Utah"]
Utah_df
```
