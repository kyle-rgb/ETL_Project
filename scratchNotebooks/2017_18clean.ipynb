```python
import pandas as pd
import numpy as np
import os 
```

```python
path18 = "2017-18/vegas.csv"
path2 = "2017-18/raw_scores.csv"
vegas18_df = pd.read_csv(path18, encoding="utf-8")
scores18_df = pd.read_csv(path2, encoding="utf-8")
vegas18_df.head()
```

```python
vegasred18_df = vegas18_df[["Team","Location", "TeamId", "GameId",
                        "PercentBet_Spread", "Average_Line_Spread", "Spread", "Result"]]
vegasred18_df
```

```python
scores18_df.head()
```

```python
redscores18_df = scores18_df[["TEAM_CITY_NAME", "TEAM_ID", "GAME_ID", "PTS",]]
redscores18_df
```

```python
scoresred18_df = redscores18_df.rename(columns={"TEAM_CITY_NAME":"Team", "TEAM_ID":"TeamId",
                                               "GAME_ID":"GameId", "PTS":"Pts"})
scoresred18_df
```

```python
merge_df = pd.merge(vegasred18_df, scoresred18_df, on=["TeamId", "GameId"])
merge_df
```

```python
grouped_df = vegasred18_df.groupby(['GameId'])
print(grouped_df)
```

```python
grouped_df.count().head()
```

```python
hawks_df = vegasred18_df.loc[vegasred18_df["Team"] == "Atlanta"]
hawks_df
```

```python
comp_df = merge_df[["Team_x", "Location", "TeamId", "GameId", "PercentBet_Spread", "Average_Line_Spread", "Spread", "Result", "Pts"]]
comp_df
```

```python
final_df = comp_df.rename(columns= {"Team_x":"Team",
                                    "Location": "Location",
                                   "TeamId":"Team Id",
                                   "GameId":"Game Id",
                                   "PercentBet_Spread":"% Bet",
                                   "Average_Line_Spread": "Avg Spread",
                                   "Spread": "Actual Spread",
                                   "Result":"Result",
                                   "Pts": "Pts"})
final_df
```

```python
Utah_df = final_df.loc[final_df["Team"] == "Utah"]
Utah_df
```
